```python
import pandas as pd
```

```python
news_df = pd.read_csv('../data/Combined_News_DJIA.csv',encoding='ISO-8859-1')
dow_df = pd.read_csv('../data/upload_DJIA_table.csv')
news_df['Date'] = pd.to_datetime(news_df['Date'])
dow_df['Date'] = pd.to_datetime(dow_df['Date'])
dow_df = dow_df.sort_values(by='Date')

dow_df['Percent_Change'] = dow_df['Close'].pct_change()
dow_df['Percent_Change'] = dow_df['Percent_Change'].fillna(0)
merged_df = pd.merge(news_df, dow_df[['Date', 'Percent_Change']], on='Date')
merged_df.dropna(inplace=True)
merged_df.head()
```

```python
train = merged_df[merged_df['Date'] < '2015-01-01']
test = merged_df[merged_df['Date'] >= '2015-01-01']
```

```python
data = train.iloc[:, 2:27]
data.replace("[^a-zA-Z]", " ", regex=True, inplace=True)

list1= [i for i in range(25)]
new_index = [str(i) for i in list1]
data.columns = new_index
data.head(5)
```

```python
import re

def clean_headline(text):
    if isinstance(text, str):
        # Remove byte-string artifacts (e.g., b'word')
        text = re.sub(r"^b['\"]", "", text)
        # Remove all punctuation
        text = re.sub(r'[^\w\s]', '', text)
        return text.lower().strip()
    return str(text).lower()

cleaned_headlines = data.iloc[:, 2:27].applymap(clean_headline)
cleaned_headlines.head(5)
```

```python
headlines = []
for row in range(0,len(cleaned_headlines.index)):
    headlines.append(' '.join(str(x)for x in cleaned_headlines.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
print("Shape of cleaned_headlines:", cleaned_headlines.shape)
print("Shape of train['Label']:", train['Label'].shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector = CountVectorizer(ngram_range=(2, 2))
traindataset=countvector.fit_transform(headlines)
```

```python
randomclassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomclassifier.fit(traindataset, train['Label'])
```

```python
test_headlines = []
for row in range(len(test)):
    test_headlines.append(' '.join(str(x) for x in test.iloc[row, 2:27]))

test_dataset = countvector.transform(test_headlines)
predictions = randomclassifier.predict(test_dataset) 
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print(accuracy_score(test['Label'], predictions))
print(classification_report(test['Label'], predictions))    
print(confusion_matrix(test['Label'], predictions))
```
